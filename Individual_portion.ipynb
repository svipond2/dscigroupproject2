```r
library(tidyverse)
```

```r
#reading and minimally tidying the players data. 
players_data<-read_tsv(url("https://raw.githubusercontent.com/zoegrames/toy_ds_project/refs/heads/main/players.csv"),col_names=FALSE,skip=1)

 players_separated <- players_data |>
   separate(col = X1, into = c("experience", "is_subscribed", "hashedEmail", "played_hours", "name","gender", "age"), sep = ",")
players_separated
```

```r
#finding the mean values of played_hours and age, the two quantitative variables
players_separated <- players_separated |>
  mutate(
    played_hours = as.numeric(played_hours),
    age = as.numeric(age))
mean_values <- players_separated |>
  summarise(
    mean_played_hours = mean(played_hours, na.rm = TRUE),
    mean_age = mean(age, na.rm = TRUE))
mean_values
```

```r
#reading and minimally tidying the sessions data
sessions_data<-read_tsv(url("https://raw.githubusercontent.com/zoegrames/toy_ds_project/refs/heads/main/sessions.csv"), col_names=FALSE, skip=1)

sessions_seperated <- sessions_data |>
  separate(col = X1, into = c("hashedEmail", "start_time", "end_time", "original_start_time", "original_end_time"), sep = ",")
sessions_seperated
```

```r
 options(repr.plot.width = 16, repr.plot.height = 9)
#Histogram to show the relationship between played hours and if a played is subscibed
played_hours_plot<-players_separated|>
    ggplot(aes( x=played_hours, fill=is_subscribed))+
    geom_histogram( stat="count",bins=30, binwidth=1)+
    labs(x="Time Played (Hours)", fill="Player is Subscribed")
played_hours_plot

#Bar plot to show the relationship between Age and Hours played
age_vs_hours_played<-players_separated|>
ggplot(aes( x=age, y=played_hours, fill=age))+
geom_bar(stat="identity", position="dodge")+
labs(y="Age(Years)", x="Player is Subscribed")
age_vs_hours_played
```
